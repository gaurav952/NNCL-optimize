```python
tags= [n for n in f_one['Final Tags']]
tags
```

```python
from django.utils.encoding import smart_str, smart_unicode
strings = [set(smart_str(i).split(' ')) for i in tags]
strings
```

```python
strings=[n for n in strings]
strings
```

```python
list_str = list()
for row in strings:
    string = ""
    for val in row:
        if string == "":
            string = string + val
        else:
            string = string + " " + val
    list_str.append(string)
list_str
```

```python
r=pd.DataFrame(f_one,columns=['barcode','title'])
r['tags'] = list_str
r
```
